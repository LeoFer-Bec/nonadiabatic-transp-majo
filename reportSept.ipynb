# Calculation of the pfaffian

```python
from IPython.display import Image
Image(filename="figs/junction2.png", width=750, height=500)
```

To determine if there are energy crossings in a topological insulator in proximity contact with two superconducting leads and three ferromagnetic islands as shown in the Fig. 1, one needs to calculate the pfaffian of the following antisymmetric matrix

\begin{equation}
\tau_y s_y \mathscr{H}_{BdG},
\end{equation} 

where the Bogoliubov-de Gennes hamiltonian of the system reads

\begin{equation}
\mathscr{H}_{BdG} = \tau_z(\mathcal{H}_{BHZ} - \mu) + \Delta(\mathbf{r})\tau_x + \mathbf{m}(\mathbf{r})\cdot\hat{s}, 
\label{bdg}
\end{equation}

with $\mathcal{H}_{BHZ}$ being the Bernevig-Hughes-Zhang hamiltonian,

\begin{equation}
\mathcal{H}_{BHZ} = A\bigl(s_zk_x\sigma_x - k_y\sigma_y\bigr) + \bigl(M-B(k_x^2+k_y^2)\bigr)\sigma_z, 
\label{bhz}
\end{equation}

and $\hat{s}$ and $\hat{\tau}$ denoting the real spin and the pseudo-spin in the Nambu space, respectively.

# Topological charge pumping
 
## Derivation of the Buttiker Pretre Thomas formula

Consider a scattering region coupled to two leads as shown in Fig. 3. Consider also a periodic potential acting in the scattering region. The form of the potential one can considered to be $U_i(r,t)=u(r)\bigl(U_{+\omega}e^{-i\omega t} + U_{-\omega}e^{+i\omega t}\bigr)$ . For simplicity one assumes that $U_i(r,t)$ is real, i.e. $U_{-\omega}=U_{+\omega}^*$. There exist a formula for the current that is pumped through the scattering region by the periodically varying potential $U_i(r,t)$. This formula, named after   Buttiker Pretre and Thomas (BPT), yields the current as the integral of the scattering matrix and its energy derivative. In the following lines I briefly described the derivation followed by BPT. 

```python
Image(filename="figs/leadsandscatter.png", width=750, height=500)
```

The starting point in the derivation of the BPT formula is the following definition of current,

\begin{equation}
I_{\alpha}(t) = \frac{e}{h} \int dE dE' \bigl[ \mathbf{a}^{\dagger}_{\alpha}(E)\mathbf{a}_{\alpha}(E')-
\mathbf{b}^{\dagger}_{\alpha}(E)\mathbf{b}_{\alpha}(E')\bigr]e^{i(E-E')t/\hbar},
\end{equation}

where $\mathbf{a}_{\alpha}^{\dagger}(E)=(a^{\dagger}_{1\alpha}(E),a^{\dagger}_{2\alpha}(E),\dots,a^{\dagger}_{m\alpha}(E),\dots)^T$ is an array of creation operators at energy $E$ in contact $\alpha$ for an incoming quantum channel $m$. On the other hand, 
$\mathbf{b}_{\alpha}^{\dagger}(E)=(b^{\dagger}_{1\alpha}(E),b^{\dagger}_{2\alpha}(E),\dots,b^{\dagger}_{m\alpha}(E),\dots)^T$
is an array of creation operators at energy $E$ and in contact $\alpha$ but for an outgoing quantum channel $m$. The treatment of the problem is within the scattering approach. Then one seeks a matrix (the S-matrix) that relates the outgoing to the incoming states, represented by the operators $\mathbf{b}^{\dagger}_{\alpha}$ and $\mathbf{a}^{\dagger}_{\alpha}$, respectively. That relationship reads $\mathbf{b}_{\alpha}=\sum_{\beta}\mathbf{s}_{\alpha\beta}\mathbf{a}_{\beta}$. However, for the case of an oscillating potential acting in the scattering region, the work of BPT decompose the outgoing states in terms of another two S-matrices, namely $\mathbf{s}_{\pm\alpha\beta}(E\pm\hbar\omega, E)$. These matrices denote the scattering amplitude of outgoing carriers appearing in lead $\alpha$ at energy $E\pm\hbar\omega$, if they were incident at energy $E$ in lead $\beta$. The addition of these extra two matrices is due to the side-bands that appear in the scattering bacause of the oscillating potential.  

\begin{eqnarray}
\mathbf{b}^{\dagger}_{\alpha}(E)&=& \mathbf{s}^{\dagger}_{\alpha\beta}(E)\mathbf{a}^{\dagger}_{\beta}(E)
+\mathbf{s}^{\dagger}_{-\alpha\beta}(E,E+\hbar\omega)eU^*_{-\omega}\mathbf{a}^{\dagger}_{\beta}(E+\hbar\omega) \\
&+& \mathbf{s}^{\dagger}_{+\alpha\beta}(E,E-\hbar\omega)eU^*_{+\omega}\mathbf{a}^{\dagger}_{\beta}(E-\hbar\omega) \\
\mathbf{b}_{\alpha}(E')&=& \mathbf{s}_{\alpha\theta}(E')\mathbf{a}_{\theta}(E')
+\mathbf{s}_{-\alpha\theta}(E',E'+\hbar\omega)eU_{-\omega}\mathbf{a}_{\theta}(E'+\hbar\omega) \\
&+& \mathbf{s}_{+\alpha\theta}(E',E'-\hbar\omega)eU_{+\omega}\mathbf{a}_{\theta}(E'-\hbar\omega)
\end{eqnarray}

Discarding the non-linear terms in the potential, the integrand of the current equation turns,

\begin{eqnarray}
&\mathbf{a}^{\dagger}_{\alpha}(E)\,\mathbf{a}_{\alpha}(E') -
\mathbf{b}^{\dagger}_{\alpha}(E)\,\mathbf{b}_{\alpha}(E') \approx
\bigl [ \delta^{\beta}_{\alpha}\delta^{\theta}_{\alpha} -
\mathbf{s}^{\dagger}_{\alpha\beta}(E)\mathbf{s}_{\alpha\theta}(E')\bigr ]\mathbf{a}^{\dagger}_{\beta}(E)\mathbf{a}_{\theta}(E')\\
&- eU_{-\omega} \bigl [ \mathbf{s}^{\dagger}_{\alpha\beta}(E)\mathbf{s}_{-\alpha\theta}(E',E'+\hbar\omega)
\mathbf{a}^{\dagger}_{\beta}(E)\mathbf{a}_{\theta}(E'+\hbar\omega)
+ \mathbf{s}^{\dagger}_{+\alpha\beta}(E,E-\hbar\omega) \mathbf{s}_{\alpha\theta}(E') 
\mathbf{a}^{\dagger}_{\beta}(E-\hbar\omega)\mathbf{a}_{\theta}(E')
\bigr ]\\
&- eU_{+\omega} \bigl [ \mathbf{s}^{\dagger}_{\alpha\beta}(E)\mathbf{s}_{+\alpha\theta}(E',E'-\hbar\omega)
\mathbf{a}^{\dagger}_{\beta}(E)\mathbf{a}_{\theta}(E'-\hbar\omega)
+ \mathbf{s}^{\dagger}_{-\alpha\beta}(E,E+\hbar\omega) \mathbf{s}_{\alpha\theta}(E') 
\mathbf{a}^{\dagger}_{\beta}(E+\hbar\omega)\mathbf{a}_{\theta}(E')
\bigr ]
\end{eqnarray}

The next step is to take the statistical average of the current density operator, keeping in mind the following avegare of the bilinear product of annihilation and creation operators,

\begin{equation}
<<a^{\dagger}_m(E) a_n(E')>> = \delta^m_n\delta(E-E')f(E),
\end{equation}

where $f(E)$ is the Fermi-Dirac function.

\begin{eqnarray}
&<<\mathbf{a}^{\dagger}_{\alpha}(E)\,\mathbf{a}_{\alpha}(E') -
\mathbf{b}^{\dagger}_{\alpha}(E)\,\mathbf{b}_{\alpha}(E')>> \approx
\mathrm{Tr}\bigl [ \delta^{\beta}_{\alpha}\delta^{\theta}_{\alpha} -
\mathbf{s}^{\dagger}_{\alpha\beta}(E)\mathbf{s}_{\alpha\theta}(E')\bigr ]\delta^{\beta}_{\theta}\delta(E-E')f(E)\\
&- eU_{-\omega} \mathrm{Tr}\bigl [ 
\mathbf{s}^{\dagger}_{\alpha\beta}(E)\mathbf{s}_{-\alpha\theta}(E',E'+\hbar\omega)f(E)
+ \mathbf{s}^{\dagger}_{+\alpha\beta}(E,E-\hbar\omega) \mathbf{s}_{\alpha\theta}(E')f(E-\hbar\omega)
\bigr ]\delta^{\beta}_{\theta}\delta(E-E'-\hbar\omega)\\
&- eU_{+\omega} \mathrm{Tr}\bigl [ 
\mathbf{s}^{\dagger}_{\alpha\beta}(E)\mathbf{s}_{+\alpha\theta}(E',E'-\hbar\omega)f(E)
+ \mathbf{s}^{\dagger}_{-\alpha\beta}(E,E+\hbar\omega) \mathbf{s}_{\alpha\theta}(E')f(E+\hbar\omega)
\bigr ]\delta^{\beta}_{\theta}\delta(E-E'+\hbar\omega)
\end{eqnarray}

Substitution of the results above back in the current equation yields 

\begin{eqnarray}
&I_{\alpha}(t) = \frac{e}{h}\int dE\Bigl \{\mathrm{Tr}\bigl[
\delta^{\beta}_{\alpha}\delta^{\theta}_{\alpha}\delta^{\beta}_{\theta} -
\mathbf{s}^{\dagger}_{\alpha\beta}(E)\mathbf{s}_{\alpha\beta}(E)\bigr ]f(E)\\
&-eU_{-\omega} \mathrm{Tr}\bigl [ \mathbf{s}^{\dagger}_{\alpha\beta}(E)\mathbf{s}_{-\alpha\beta}(E-\hbar\omega,E) f(E)
+ \mathbf{s}^{\dagger}_{+\alpha\beta}(E,E-\hbar\omega) \mathbf{s}_{\alpha\beta}(E-\hbar\omega) f(E-\hbar\omega) \bigr]
e^{i\omega t}\\
&-eU_{+\omega} \mathrm{Tr}\bigl [ \mathbf{s}^{\dagger}_{\alpha\beta}(E)\mathbf{s}_{+\alpha\beta}(E+\hbar\omega,E) f(E)
+ \mathbf{s}^{\dagger}_{-\alpha\beta}(E,E+\hbar\omega) \mathbf{s}_{\alpha\beta}(E+\hbar\omega) f(E+\hbar\omega) \bigr]
e^{-i\omega t}\Bigr \}.
\end{eqnarray}

Further simplifications in the above expression are achieved by the fact that the outgoing states obey the same statistics as the incoming states, i.e. $\{b^{\dagger}_{m\alpha}(E),b_{n\beta}(E')\}=\delta^m_n \delta^{\alpha}_{\beta}\delta(E-E')$. Then in first order of the periodic potential ($U_{-\omega}$ and $U_{+\omega}$), the following equations can be obtained straightforwardly,

\begin{eqnarray}
&\mathbf{s}_{\alpha\beta}^{\dagger}(E)\mathbf{s}_{\alpha\beta}(E') = \mathbf{1} \\
&\mathbf{s}^{\dagger}_{\alpha\beta}(E)\mathbf{s}_{-\alpha\beta}(E-\hbar\omega,E)
+ \mathbf{s}^{\dagger}_{+\alpha\beta}(E,E-\hbar\omega) \mathbf{s}_{\alpha\beta}(E-\hbar\omega)=0 \\
&\mathbf{s}^{\dagger}_{\alpha\beta}(E)\mathbf{s}_{+\alpha\beta}(E+\hbar\omega,E) 
+ \mathbf{s}^{\dagger}_{-\alpha\beta}(E,E+\hbar\omega) \mathbf{s}_{\alpha\beta}(E+\hbar\omega)=0
\end{eqnarray}

Substitution of the results above back in the current equation yields 

\begin{eqnarray}
&I_{\alpha}(t) = \frac{e}{h}\int dE\Bigl \{ 
-eU_{-\omega} \mathrm{Tr}\bigl [ \mathbf{s}^{\dagger}_{\alpha\beta}(E)\mathbf{s}_{-\alpha\beta}(E-\hbar\omega,E) 
\bigl( f(E) - f(E-\hbar\omega) \bigr) \bigr] e^{i\omega t}\\
&-eU_{+\omega} \mathrm{Tr}\bigl [ \mathbf{s}^{\dagger}_{\alpha\beta}(E)\mathbf{s}_{+\alpha\beta}(E+\hbar\omega,E)
\bigl( f(E) - f(E+\hbar\omega) \bigr)  \bigr] e^{-i\omega t}\Bigr \}.
\end{eqnarray} 

The work of BPT take only the term proportional to $U_{+\omega}$ from the above current expression, 

\begin{equation}
I_{\alpha}(t) = -\Bigl \{\frac{e^2}{h}\int dE \, 
\mathrm{Tr}\bigl [ \mathbf{s}^{\dagger}_{\alpha\beta}(E)\mathbf{s}_{+\alpha\beta}(E+\hbar\omega,E)
\bigl( f(E) - f(E+\hbar\omega) \bigr)  \bigr] e^{-i\omega t}\Bigr \}U_{+\omega}.
\end{equation} 

## An example of charge pumping

The equations that I am going to solve appear below

\begin{equation}
\Delta Q = -\frac{i}{2\pi}\int^T_0 dt \int_{-\infty}^{\infty} dE\rho'(E)\,\Xi_{jj}(E,t),
\end{equation}

where $\rho'(E)$ is the derivative of the Fermi-Dirac function with respect to energy and $\,\Xi(E,t)=i\frac{dS(E,t)}{dt}S^*(E,t)$, with $S(E,t)$ being the scattering matrix. For the system of Fig. 1 the S matrix reads,

\begin{equation}
S(E,t)= \left( \begin{array}{cc} r_{ee}(E,t) & r_{eh}(E,t)\\ r_{he}(E,t) & r_{hh}(E,t) \end{array}\right)
\end{equation}

where $r_{ee}$ and $r_{eh}$ are the normal and Andreev reflection coefficients, respectively. The total charge pump into the system can be decomposed in two terms, namely the electron and hole charge pumpings,

\begin{equation}
\Delta Q_e = \frac{1}{2\pi}\int^T_0 dt \int_{-\infty}^{\infty} dE\rho'(E)\Bigl(\frac{\partial r_{ee}}{\partial t}r_{ee}^* + \frac{\partial r_{eh}}{\partial t}r_{he}^* \Bigr),
\end{equation}

\begin{equation}
\Delta Q_h = \frac{1}{2\pi}\int^T_0 dt \int_{-\infty}^{\infty} dE\rho'(E)\Bigl(\frac{\partial r_{hh}}{\partial t}r_{hh}^* + \frac{\partial r_{he}}{\partial t}r_{eh}^* \Bigr).
\end{equation}

## The BdG hamiltonian

```python
Image(filename="figs/qpumping1d.png", width=750, height=500)
```

The one dimensional hamiltonian where superconducting and magnetic islands are in proximately contact with the helical states of the quantum spin hall effect reads,

\begin{equation}
\mathscr{H}_{BdG} = \left( 
\begin{array}{cccc} 
v_{F}p_x -\mu & m_{||}(x)e^{-i\alpha}& \Delta(x) & \\ 
m_{||}(x)e^{i\alpha} & -v_{F}p_x -\mu&  & \Delta(x)\\ 
\Delta(x)  & & -v_{F}p_x +\mu & m_{||}(x)e^{-i\alpha} \\
& \Delta(x) & m_{||}(x)e^{-i\alpha} & v_{F}p_x +\mu \end{array}\right)
\end{equation}

### The QSH region

\begin{equation}
\mathscr{H}_{BdG} = \left( 
\begin{array}{cccc} 
v_{F}p_x -\mu_{QSH} &  &  & \\
 & -v_{F}p_x -\mu_{QSH} &  & \\
 &  & -v_{F}p_x +\mu_{QSH} &  \\
 &  &  & v_{F}p_x +\mu_{QSH} 
\end{array}\right)
\end{equation}

The four self-states of the hamiltonian above have the linear momenta,

\begin{equation}
k = \pm \frac{\mu_{QSH}\pm E}{v_F},
\end{equation}

where $\mu_{QSH}$ is the chemical potential in the topological insulator. We consider incident states from $x=-\infty$ to $x=-L$ and normal and Andreev 
reflection states,

\begin{equation}
\psi_{QSH}(x) =  \left ( 
\begin{array}{c} 
1 \\ 0 \\ 0 \\ 0 
\end{array}\right)e^{i\frac{E + \mu_{QSH}}{v_F}x} +
r_{ee} \left ( 
\begin{array}{c} 
0 \\ 1 \\ 0 \\ 0 
\end{array}\right)e^{-i\frac{E + \mu_{QSH}}{v_F}x} +
r_{eh} \left ( 
\begin{array}{c}
0 \\ 0 \\ 1 \\ 0 \end{array}\right)e^{i\frac{\mu_{QSH}-E}{v_F}x} 
\end{equation}

### The magnetic region

\begin{equation}
\mathscr{H}_{BdG} = \left( 
\begin{array}{cccc} 
v_{F}p_x -\mu_{FI} & m_{||}e^{-i\alpha}&  & \\
m_{||}e^{i\alpha} & -v_{F}p_x -\mu_{FI} &  & \\
& & -v_{F}p_x +\mu_{FI} & m_{||}e^{-i\alpha} \\
&  & m_{||}e^{-i\alpha} & v_{F}p_x +\mu_{FI} 
\end{array}\right)
\label{hamferile}
\end{equation}

The four self-states of the hamiltonian above have the linear momenta,

\begin{equation}
k = \pm i\frac{\sqrt{m_{||}^2 - (E\pm \mu_{FI})^2}}{v_F},
\end{equation}

where $\mu_{FI}$ is the chemical potential in the ferromagnetic island, and for simplicity we set it to zero. In this case the self-states 
of the hamiltonian read

\begin{equation}
\psi_{FI}(x) = b \left ( \begin{array}{c} e^{i\theta_m}e^{-i\alpha} \\ 1 \\ 0 \\ 0 \end{array}\right)e^{\frac{-\sqrt{m_{||}^2-E^2}}{v_F}x} +
b' \left ( \begin{array}{c} e^{-i\theta_m}e^{-i\alpha} \\ 1 \\ 0 \\ 0 \end{array}\right)e^{\frac{\sqrt{m_{||}^2-E^2}}{v_F}x} +
d \left ( \begin{array}{c} 0 \\ 0 \\ e^{i\theta_m}e^{-i\alpha} \\ 1 \end{array}\right)e^{\frac{\sqrt{m_{||}^2-E^2}}{v_F}x} +
d' \left ( \begin{array}{c} 0 \\ 0 \\ e^{-i\theta_m}e^{-i\alpha} \\ 1 \end{array}\right)e^{-\frac{\sqrt{m_{||}^2-E^2}}{v_F}x}, 
\end{equation}

where $\cos{\theta_m}=E/m_{||}$.

### The superconducting region

\begin{equation}\mathscr{H}_{BdG} = \left( 
\begin{array}{cccc} 
v_{F}p_x -\mu & & \Delta_0 & \\  
& -v_{F}p_x -\mu&  & \Delta_0\\ 
\Delta_0  & & -v_{F}p_x +\mu & \\ 
 & \Delta_0 &  & v_{F}p_x +\mu
\end{array}\right)\end{equation}

The four self-states of the above hamiltonian have the linear momenta,

\begin{equation}
k = \pm\frac{i\sqrt{\Delta_0^2-E^2} \pm \mu_{SC}}{v_F}.
\end{equation}

We consider below only the states propagating to the right, i.e. those states with the overall sign positive. They read

\begin{equation}
\psi_{SC}(x) = c \left ( 
\begin{array}{c} 
e^{i\beta} \\ 0 \\ 1 \\ 0
\end{array}
\right)e^{i\frac{\mu_{SC}}{v_F}x}\,e^{\frac{-\sqrt{\Delta_{0}^2-E^2}}{v_F}x} +
f \left ( 
\begin{array}{c}
0 \\ e^{-i\beta} \\ 0 \\ 1
\end{array}\right)e^{-i\frac{\mu_{SC}}{v_F}x}\,e^{\frac{-\sqrt{\Delta_{0}^2-E^2}}{v_F}x}, 
\end{equation}

where $\cos{\beta}= E/\Delta_0$.

### The reflection coefficients

Imposing the condition that the wave function is continuous at the interfaces, 

\begin{equation}
\psi_{QSH}(0)=\psi_{FI}(0), \quad and \quad  \psi_{FI}(L)=\psi_{SC}(L),  
\end{equation}

one obtains a set of linear equations for the coefficients of the wave functions in the three regions describe above. The set of equations read

\begin{equation}
\left(
\begin{array}{cccccccc}
e^{i(\theta_m-\alpha)} & e^{-i(\theta_m+\alpha)} & & & & & &  \\
1 & 1 & -1 & & & & & &\\
e^{i(\theta_m-\alpha+iqL)} & e^{-i(\theta_m+\alpha+iqL)} & & -e^{i\bigl(\beta +(\mu_{SC}L)/v_F+ipL\bigr)} & & & & \\
e^{-qL} & e^{qL} & & & -e^{-i\bigl(\beta +(\mu_{SC}L)/v_F-ipL\bigr)} & & & \\
 & & & & & e^{i(\theta_m-\alpha)} & e^{-i(\theta_m+\alpha)} & -1 \\
 & & & & & 1 & 1 & \\
 & & & -e^{i\bigl(\frac{\mu_{SC}}{v_F}+ip\bigr)L} & & e^{i(\theta_m-\alpha-iqL)} & e^{-i(\theta_m+\alpha-iqL)} & \\
 & & & & -e^{-i\bigl(\frac{\mu_{SC}}{v_F}-ip\bigr)L} & e^{qL} & e^{-qL} &  \\
\end{array}
\right)
\left(
\begin{array}{c}
b \\
b'\\
r_{ee}\\
c \\
f \\
d \\
d'\\
r_{eh}
\end{array}
\right)
= 
\left(
\begin{array}{c}
1 \\
0 \\
0 \\
0 \\
0 \\
0 \\
0 \\
0
\end{array}
\right)
\end{equation}

The work of X-J Luo et al. calculated the reflection coefficients, they are:
 
\begin{equation}
r_{ee} = \frac{e^{i(\alpha+\theta_m)}\bigl(-1 + e^{2k_{FI}L}\bigr) 
\bigl(-e^{2i\beta} + e^{2i\theta_m} - e^{2k_{FI}L} \bigr)}{e^{2i\beta}}
\end{equation}
